```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# عرض المخططات داخل الدفتر
%matplotlib inline

# إعدادات جمالية للمخططات
sns.set(style="darkgrid")

# تحميل البيانات
df = pd.read_csv('/content/owid-covid-data.csv')

# عرض أول 5 صفوف
df.head()
```

```python
# استخراج فقط الصفوف الخاصة بالأردن
jordan_df = df[df['location'] == 'Jordan']

# عرض أول 5 صفوف من بيانات الأردن
jordan_df.head()
```

```python
# تحويل التاريخ إلى صيغة datetime
jordan_df['date'] = pd.to_datetime(jordan_df['date'])

# التأكد من النوع الجديد
jordan_df.dtypes[['date']]
```

```python
plt.figure(figsize=(12,6))
plt.plot(jordan_df['date'], jordan_df['total_cases'], color='blue')
plt.title('Total COVID-19 Cases in Jordan Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(jordan_df['date'], jordan_df['new_cases'], color='orange')
plt.title('Daily New COVID-19 Cases in Jordan')
plt.xlabel('Date')
plt.ylabel('New Cases Per Day')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(jordan_df['date'], jordan_df['total_deaths'], color='red')
plt.title('Total COVID-19 Deaths in Jordan Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.grid(True)
plt.show()
```

```python
# معرفة عدد القيم المفقودة في كل عمود
jordan_df.isnull().sum()
```

```python
# حذف الصفوف التي تحتوي على حالات جديدة أو وفيات جديدة مفقودة
jordan_df_cleaned = jordan_df.dropna(subset=['new_cases', 'new_deaths'])

# تأكد بعد التنظيف
jordan_df_cleaned.isnull().sum()
```

```python
# حذف الأعمدة الفارغة
jordan_df_cleaned = jordan_df_cleaned.drop(columns=[
    'excess_mortality',
    'excess_mortality_cumulative',
    'excess_mortality_cumulative_absolute',
    'excess_mortality_cumulative_per_million'
])
```

```python
# استخراج اليوم الذي سجلت فيه أعلى عدد من الحالات الجديدة
max_cases_day = jordan_df_cleaned.loc[jordan_df_cleaned['new_cases'].idxmax()]
print(" أعلى يوم إصابات:", max_cases_day['date'])
print(" عدد الحالات:", int(max_cases_day['new_cases']))
```

```python
# آخر قيمة متوفرة للحالات والوفيات
latest_data = jordan_df_cleaned.iloc[-1]

# حساب النسبة المئوية
death_rate = (latest_data['total_deaths'] / latest_data['total_cases']) * 100
print(f" نسبة الوفيات للحالات الكلية في الأردن: {death_rate:.2f}%")
```

```python
plt.figure(figsize=(12,6))
plt.plot(jordan_df_cleaned['date'], jordan_df_cleaned['new_cases'], label='New Cases', color='orange')
plt.plot(jordan_df_cleaned['date'], jordan_df_cleaned['new_deaths'], label='New Deaths', color='red')
plt.title('New Cases vs New Deaths in Jordan')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig = px.line(jordan_df_cleaned, x='date', y='total_cases',
              title=' Total COVID-19 Cases in Jordan (Interactive)',
              labels={'date': 'Date', 'total_cases': 'Total Cases'})
fig.show()
```
